```python
import pandas as pd
import plotly.express as px

# Clean and prepare the country data
country_series = df['country'].dropna().str.split(', ')
country_flat = country_series.explode().str.strip()
country_data = country_flat.value_counts().reset_index()
country_data.columns = ['Country', 'Total Titles']

# Create an interactive choropleth map
fig = px.choropleth(
    country_data,
    locations='Country',
    locationmode='country names',
    color='Total Titles',
    color_continuous_scale=px.colors.sequential.Magma,
    hover_name='Country',
    title='🌍 Netflix Content Distribution by Country',
)

# Beautify layout
fig.update_layout(
    title_font_size=24,
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='natural earth'
    ),
    font=dict(
        family="Arial",
        size=14,
        color="#4f4f4f"
    ),
    margin=dict(l=0, r=0, t=50, b=0),
    coloraxis_colorbar=dict(
        title="Total Titles",
        tickfont=dict(size=12),
        titlefont=dict(size=14)
    )
)

fig.show()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Prepare genre data
df['listed_in'] = df['listed_in'].fillna('')
genres = df['listed_in'].str.split(', ')
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(genres)

# Optional: include country or director info too
# This example clusters only on genres
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(genre_encoded)

# Visualize with PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(genre_encoded)

plt.figure(figsize=(10, 6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=df['cluster'], cmap='rainbow', alpha=0.5)
plt.title("Clustering of Netflix Titles by Genre")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```

```python
#Network Graph: Directors and Actors (Co-working Relationships)
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

for _, row in df.dropna(subset=['director', 'cast']).iterrows():
    director = row['director']
    actors = row['cast'].split(', ')
    for actor in actors:
        G.add_edge(director.strip(), actor.strip())

# Filter for most connected nodes to avoid clutter
degree_dict = dict(G.degree())
top_nodes = sorted(degree_dict, key=degree_dict.get, reverse=True)[:50]
subgraph = G.subgraph(top_nodes)

plt.figure(figsize=(12, 8))
nx.draw(subgraph, with_labels=True, node_size=500, node_color='lightblue', edge_color='gray', font_size=8)
plt.title("Network of Directors and Actors")
plt.show()
```

```python
# Recommendation Engine Based on Cast/Director Similarity
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Combine metadata into a single string
df['combined'] = df['director'].fillna('') + ' ' + df['cast'].fillna('')

# Convert to count matrix
vectorizer = CountVectorizer().fit_transform(df['combined'])
similarity = cosine_similarity(vectorizer)

# Recommend similar titles
def recommend(title):
    index = df[df['title'] == title].index[0]
    scores = list(enumerate(similarity[index]))
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:6]
    for i in sorted_scores:
        print(df.iloc[i[0]]['title'])

# Example use
recommend("Narcos")
```

```python
from textblob import TextBlob

# Ensure 'description' column is not null
df['description'] = df['description'].fillna("")

# Apply sentiment polarity scoring
df['sentiment'] = df['description'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Visualize sentiment polarity
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(df['sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title("Sentiment Distribution of Netflix Descriptions")
plt.xlabel("Polarity (Negative to Positive)")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```
